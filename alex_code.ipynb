```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#DoorDash: https://www.macrotrends.net/stocks/charts/DASH/doordash/income-statement
#UberEats: https://investor.uber.com/financials/default.aspx
#DeliveryHero: https://ir.deliveryhero.com/financial-reports-and-presentations/
#GrubHub: https://www.businessofapps.com/data/grubhub-statistics/
#Instacart: https://www.businessofapps.com/data/instacart-statistics/
#HelloFresh: https://www.statista.com/statistics/655060/hellofresh-net-revenue-worldwide/


doorDash_rev = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "Revenue (Billions)": [0.291, 0.885, 2.886, 4.888, 6.583],
}

uberEats_rev={
        "Year": [2017, 2018, 2019, 2020, 2021, 2022],
        "Revenue (Billions)": [0.355, 0.772, 1.400, 3.904, 8.362, 10.901]
}

deliveryHero_rev = {
    "Year": [2017, 2018, 2019, 2020, 2021, 2022],
    "Revenue (Billions)": [0.597, 0.731, 1.360, 2.716, 6.434, 9.423],
}

grubHub_rev={
        "Year": [2017, 2018, 2019, 2020, 2021, 2022],
        "Revenue (Billions)": [0.683, 1.007, 1.312, 2.026, 2.362, 2.482]
}

instaCart_rev={
        "Year": [2017, 2018, 2019, 2020, 2021, 2022],
        "Revenue (Billions)": [0.300, 0.525, 0.735, 1.500, 1.800, 2.500]
}

helloFresh_rev = {
    "Year": [2017, 2018, 2019, 2020, 2021, 2022],
    "Revenue (Billions)": [0.994, 1.405, 1.987, 4.119, 6.582, 8.355],
}


plt.plot(doorDash_rev["Year"], doorDash_rev["Revenue (Billions)"], label="DoorDash")
plt.plot(uberEats_rev["Year"], uberEats_rev["Revenue (Billions)"], label = "UberEats")
plt.plot(deliveryHero_rev["Year"], deliveryHero_rev["Revenue (Billions)"], label="DeliveryHero")
plt.plot(grubHub_rev["Year"], grubHub_rev["Revenue (Billions)"], label = "GrubHub")
plt.plot(instaCart_rev["Year"], instaCart_rev["Revenue (Billions)"], label = "Instacart")
plt.plot(helloFresh_rev["Year"], helloFresh_rev["Revenue (Billions)"], label="HelloFresh")
plt.xticks(np.arange(2017, 2023, 1))
plt.yticks(np.arange(0, 13, 2))
plt.xlabel("Year")
plt.ylabel("Revenue (Billions USD)")
plt.title("Food Delivery Companies' Annual Revenue")
plt.legend(loc="upper left")
plt.show()

```
